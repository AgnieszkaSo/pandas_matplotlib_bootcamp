## Wypełnianie pola pomiędzy liniami

```python
x = np. linspace(0, 2, 100)
```

```python
y1 = x
y2 = x**2
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (8, 5)) 
plt.title('Funkcje x i x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x, y1, linestyle = '--', color = '#444444', label = 'x', linewidth = 2) 
plt.plot(x, y2, linestyle = '-', color = '#33c7ff', label = 'x^2', linewidth = 2) 
plt.legend(loc = "upper left")
plt.grid(True)
plt.show()
```

dodawanie wypełnienia poniżej/powyżej zadanego poziomu

```python
plt.style.use('ggplot')
plt.figure(figsize = (8, 5)) 
plt.title('Funkcje x i x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x, y1, linestyle = '--', color = '#444444', label = 'x', linewidth = 2) 
plt.plot(x, y2, linestyle = '-', color = '#33c7ff', label = 'x^2', linewidth = 2) 

poziom = 1
plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1")

plt.legend(loc = "upper left")
plt.grid(True)
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize = (8, 5)) 
plt.title('Funkcje x i x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x, y1, linestyle = '--', color = '#444444', label = 'x', linewidth = 2) 
plt.plot(x, y2, linestyle = '-', color = '#33c7ff', label = 'x^2', linewidth = 2) 

poziom = 1
plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1, x^2 > 1", where = (y2 > 1))
plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1, x^2 <= 1", where = (y2 <= 1))


plt.legend(loc = "upper left")
plt.grid(True)
plt.show()
```
