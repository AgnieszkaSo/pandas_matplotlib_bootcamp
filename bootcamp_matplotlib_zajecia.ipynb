## Zadanie 1

Dla danych x narysuj funkcje: f(x) = x, f(x) = x^2, f(x) = x^3. 

```python
x = np.linspace(-1.5, 1.5, 100)
```

```python
x
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
#plt.legend(loc = "lower right") # lower left, lower right, upper left, upper right, bądź wartośći liczbowe
```

```python
plt.style.use("ggplot")
plt.figure(figsize = (7, 7))
plt.title("Funkcje x, x^2 i x^3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y1, linestyle = "--", linewidth = 2, label = "x")
plt.plot(x, y2, linestyle = "-", linewidth = 2, label = "x^2")
plt.plot(x, y3, linestyle = ":", linewidth = 2, label = "x^3")
plt.xlim([-3.5, 3.5]) # zakres osi OX
plt.ylim([-3.5, 3.5]) # zakres osi OY
plt.axhline(y = 0, color = "#444444") # dodajemy linię poziomą dla y =0 (horizontal)
plt.axvline(x = 0, color = "#444444") # dodajemy linię pionową dla x =0 (vertical)
plt.legend()
plt.grid(True)
plt.show()
```

## Wypełnianie pola pomiędzy liniami

```python
x = np.linspace(0, 2, 100)
```

```python
y1 = x
y2 = x**2
```

```python
plt.style.use("ggplot")
plt.figure(figsize = (8, 5))
plt.title("Funkcje x i x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y1, linestyle = "--", linewidth = 2, label = "x")
plt.plot(x, y2, linestyle = "-", linewidth = 2, label = "x^2")
plt.grid(True)
plt.show()
```

dodawanie wypełnienia poniżej/powyżej zadanego poziomu

```python
plt.figure(figsize = (8, 5))
plt.title("Funkcje x i x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y1, linestyle = "--", linewidth = 2, label = "x")
plt.plot(x, y2, linestyle = "-", linewidth = 2, label = "x^2")
poziom = 1
plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8, 5))
plt.title("Funkcje x i x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y1, linestyle = "--", linewidth = 2, label = "x")
plt.plot(x, y2, linestyle = "-", linewidth = 2, label = "x^2")
poziom = 1

plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1, x^2 > 1", where = (y2>1)) # alpha to przejrzystość
plt.fill_between(x, poziom, y2, alpha = 0.25, label = "pole pomiędzy x^2 i 1, x^2 <= 1", where = (y2 <= 1))

plt.grid(True)
plt.legend()
plt.show()
```

## Zadanie 2

Narysuj podobny wykres, tylko zamiast wypełnienia pomiędzy x^2 i 1, dodaj wypełnienie pomiędzy x^2 i x. 

```python
plt.figure(figsize = (8, 5))
plt.title("Funkcje x i x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y1, linestyle = "--", linewidth = 2, label = "x")
plt.plot(x, y2, linestyle = "-", linewidth = 2, label = "x^2")

plt.fill_between(x, y1, y2, alpha = 0.25, label = "pole pomiędzy x^2 i x, x^2 > x", where = (y2>y1)) # alpha to przejrzystość, intensywność koloru
plt.fill_between(x, y1, y2, alpha = 0.25, label = "pole pomiędzy x^2 i x, x^2 <= x", where = (y2 <= y1))

plt.grid(True)
plt.legend()
plt.show()
```
